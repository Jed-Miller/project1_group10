```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
outputpath = "figures/HighLowSnow_bargraph.png"
outputpath1 = "figures/2015_corresponding_bargraph.png"
outputpath2 = "figures/2019_corresponding_bargraph.png"
outputpath3 = "figures/2015_following_bargraph.png"
outputpath4 = "figures/2019_following_bargraph.png"
path = "data/full_duration.csv"
county_fire_duration = pd.read_csv(path)
county_fire_duration.head()
```

```python
#Isolating the years for the specific bar graphs
cfd_2015 = county_fire_duration.loc[county_fire_duration['Year'] == 2015]
cfd_2016 = county_fire_duration.loc[county_fire_duration['Year'] == 2016]
cfd_2019 = county_fire_duration.loc[county_fire_duration['Year'] == 2019]
cfd_2020 = county_fire_duration.loc[county_fire_duration['Year'] == 2020]
```

```python
x = cfd_2015['County']
y = cfd_2015['Avg. Fire Duration (Days)']


plt.figure(figsize =(20,10))
plt.bar(x,y,width = 1, linewidth = 1,label = '2015 Low Snow ~ Avg. 23.51 in.')
plt.xlabel('County', fontsize = 22)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.ylim(0,20)
plt.ylabel('Avg. Fire Duration (Days)', fontsize = 22)
plt.legend(loc = 'upper left')
plt.title("Avg. Fire Duration for Min Snowpack Year (2015) (Corresponding Fire Season)", fontsize = 22)
plt.savefig(outputpath1)
plt.show()
```

```python
x1 = cfd_2019['County']
y1 = cfd_2019['Avg. Fire Duration (Days)']

plt.figure(figsize =(20,10))
plt.bar(x1,y1,width = 1, linewidth = 1,color= 'red', label = '2019 High Snow ~Avg. 84.85 in.')
plt.xlabel('County', fontsize = 22)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.ylim(0,20)
plt.ylabel('Avg. Fire Duration (Days)', fontsize = 22)
plt.legend(loc = 'upper left')
plt.title("Avg. Fire Duration for Max Snowpack Year (2019) (Corresponding Fire Season)", fontsize = 22)
plt.savefig(outputpath2)
plt.show()
```
