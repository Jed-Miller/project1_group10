```python
#Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
#Call in data and create initial dataframe
outputpath = "figures/avg_fire_duration_county.png"
path = "data/full_duration.csv"
county_fire_duration = pd.read_csv(path)
county_fire_duration.head()
```

```python
#Filter out fire duration data that was inputted inaccurately (i.e. Cal Fire Source Data is inaccurate.)
county_fire_duration_filtered = county_fire_duration[county_fire_duration["Median Fire Duration (Days)"] <= 70]

```

```python
#Create line graph for Avg Fire Duration per County (with snow depth)
amador_data = county_fire_duration_filtered.loc[county_fire_duration_filtered["County"] == "Amador"]
butte_data = county_fire_duration_filtered.loc[county_fire_duration_filtered["County"] == "Butte"]
calaveras_data = county_fire_duration_filtered.loc[county_fire_duration_filtered["County"] == "Calaveras"]
el_dorado_data = county_fire_duration_filtered.loc[county_fire_duration_filtered["County"] == "El Dorado"]
nevada_data = county_fire_duration_filtered.loc[county_fire_duration_filtered["County"] == "Nevada"]
placer_data = county_fire_duration_filtered.loc[county_fire_duration_filtered["County"] == "Placer"]
sacramento_data = county_fire_duration_filtered.loc[county_fire_duration_filtered["County"] == "Sacramento"]
sierra_data = county_fire_duration_filtered.loc[county_fire_duration_filtered["County"] == "Sierra"]
yuba_data = county_fire_duration_filtered.loc[county_fire_duration_filtered["County"] == "Yuba"]
alpine_data = county_fire_duration_filtered.loc[county_fire_duration_filtered["County"] == "Alpine"]

fig, ax = plt.subplots(figsize=(20,10))

butte_data.plot(x = "Year", y = "Mean Snow Depth (in)", ax = ax, linewidth = 7)
amador_data.plot(x = "Year", y = "Avg. Fire Duration (Days)", ax = ax, secondary_y = True)
butte_data.plot(x = "Year", y = "Avg. Fire Duration (Days)", ax = ax, secondary_y = True)
calaveras_data.plot(x = "Year", y = "Avg. Fire Duration (Days)", ax = ax, secondary_y = True)
el_dorado_data.plot(x = "Year", y = "Avg. Fire Duration (Days)", ax = ax, secondary_y = True)
nevada_data.plot(x = "Year", y = "Avg. Fire Duration (Days)", ax = ax, secondary_y = True)
placer_data.plot(x = "Year", y = "Avg. Fire Duration (Days)", ax = ax, secondary_y = True)
sacramento_data.plot(x = "Year", y = "Avg. Fire Duration (Days)", ax = ax, secondary_y = True)
sierra_data.plot(x = "Year", y = "Avg. Fire Duration (Days)", ax = ax, secondary_y = True)
yuba_data.plot(x = "Year", y = "Avg. Fire Duration (Days)", ax = ax, secondary_y = True)
alpine_data.plot(x = "Year", y = "Avg. Fire Duration (Days)", ax = ax, secondary_y = True)

plt.xlabel("Year", fontsize = 22)
plt.title("Avg. Fire Duration by County", fontsize = 22)
plt.savefig(outputpath)
plt.show()

```
