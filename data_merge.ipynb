# TIME TO MERGE

```python
import pandas as pd
```

```python
snow_data_path = "data/snow_data.csv"
yearly_fire_path = "data/yearly_fire_data.csv"
county_fire_path = "data/county_fire_data.csv"
duration_fire_path = "data/duration_fire_data.csv"

snow_data = pd.read_csv(snow_data_path)
yearly_fire = pd.read_csv(yearly_fire_path)
county_fire = pd.read_csv(county_fire_path)
duration_fire = pd.read_csv(duration_fire_path)
snow_data
```

```python
yearly_fire = yearly_fire.drop(columns = ["Avg. Duration (Hours)", "Median Duration (Hours)"])
yearly_fire = yearly_fire.rename(columns = {
    "Avg. Acres" : "Avg. Acres Burned",
    "Median Acres" : "Median Acres Burned",
    "Avg. Duration (Days)" : "Avg. Fire Duration (Days)",
    "Median Duration (Days)" : "Median Fire Duration (Days)"
})

duration_fire = duration_fire.drop(columns = ["Avg. Duration (Hours)", "Median Duration (Hours)"])
duration_fire = duration_fire.rename(columns = {
    "Avg. Acres" : "Avg. Acres Burned",
    "Median Acres" : "Median Acres Burned",
    "Avg. Duration (Days)" : "Avg. Fire Duration (Days)",
    "Median Duration (Days)" : "Median Fire Duration (Days)"
})
```

```python
year_data = pd.merge(yearly_fire, snow_data, on = "Year", how = "left")
year_data = year_data.drop(columns = ["Season"])
year_data = year_data.round(decimals = 2)
# Format for final tables
year_data
```

```python
county_data = pd.merge(county_fire, snow_data, on = "Year", how = "left")
county_data = county_data.drop(columns = ["Season"])
county_data = county_data.round(decimals = 2)
# Format for final tables
county_data
```

```python
duration_data = pd.merge(duration_fire, snow_data, on = "Year", how = "left")
duration_data = duration_data.drop(columns = ["Season"])
duration_data = duration_data.round(decimals = 2)
# Format for final tables
duration_data
```

```python
yearly_data_output = "data/full_yearly.csv"
county_data_output = "data/full_county.csv"
duration_data_output = "data/full_duration.csv"

year_data.to_csv(yearly_data_output)
county_data.to_csv(county_data_output)
duration_data.to_csv(duration_data_output)
```
